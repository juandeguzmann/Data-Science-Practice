```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math

## Read in Data
data = pd.read_csv("all_data.csv")
print(data.head())
```

```python
unique_countries = pd.unique(data.Country)

for country in unique_countries:
    data_unique_country = data[data.Country == country]
    plt.plot(data_unique_country.Year, data_unique_country['Life expectancy at birth (years)'])

plt.legend(unique_countries)
plt.ylabel('Life Expectancy')
plt.xlabel('Year')
plt.show()
plt.clf()
```

```python
for country in unique_countries:
    data_unique_country = data[data.Country == country]
    plt.plot(data_unique_country.Year, data_unique_country['GDP'])

plt.legend(unique_countries)
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
plt.clf()
```

```python
for country in unique_countries:
    data_unique_country = data[data.Country == country]
    plt.scatter( data_unique_country['Life expectancy at birth (years)'], data_unique_country.GDP)

plt.legend(unique_countries)
plt.ylabel('GDP')
plt.xlabel('Life Expectancy')
plt.show()
plt.clf()
```

```python
for country in unique_countries:
    data_unique_country = data[data.Country == country]
    average_life_expectancy = np.mean(data_unique_country['Life expectancy at birth (years)'])
    print('Average life expectancy in ' + str(country) + ' is: ' + str(average_life_expectancy) + ' Years old')
```

```python
for country in unique_countries:
    data_unique_country = data[data.Country == country]
    plt.hist(data_unique_country['Life expectancy at birth (years)'], alpha = 0.5 , density=True, range= (70,90), bins= 10)
plt.legend(unique_countries)
plt.ylabel('Density')
plt.xlabel('life Expectancy (Years)')
plt.show()
plt.clf()
```
